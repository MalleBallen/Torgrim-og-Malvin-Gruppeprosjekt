```python
class ODESolver:
    def __init__(self,
                 y,
                 t0,
                 tf,
                 f,
                 h, 
                 method='euler'):
        self.y = y
        self.f_i = f(y, t0)
        self.t0 = t0
        self.tf = tf
        self.f = f
        self.h = h

        self.method = method.lower()

        
    def euler_vector_step(y, t, f, h):
        dy = f(y, t)

        y_next = y + h * dy
        return y_next

    def runge_kutta_2nd_step(y, t, f, h):
        k1 = h*f(y, t)
        k2 = h*f(y + h * k1 / 2, t + h / 2)
        y_next = y + k2

        return y_next

    def runge_kutta_4th_step(y, t, f, h):
        k1 = h*f(y, t)
        k2 = h*f(y + h * k1 / 2, t + h / 2)
        k3 = h*f(y + h * k2 / 2, t + h / 2)
        k4 = h*f(y + h * k3, t + h)

        y_next = y + (k1 + 2 * k2 + 2 * k3 + k4)/6
        return y_next



```
